```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
import warnings
warnings.filterwarnings('ignore')
```

### Добавление полиномиальных признаков

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.base import BaseEstimator, TransformerMixin
```

### Разделение на обучение и тест. Обработка данных.

```python
X_train = data[feature_cols]
y_train = data[target_col]
```

### Напишем собственные классы для обработки данных

```python
# метод замены пробелов на 0 и преобразование типа для TotalSpent

class FeatureTransformer(BaseEstimator, TransformerMixin): 
    def __init__(self):
        return None
    def fit(self, X, y = None):
        return self
    def transform(self, X):
        # в датасете присутствуют пустые ячейки в виде строки пробела
        X = X.replace(' ', '0') # заменим все пробелы (пустые ячейки) на нуль
        X = X.astype({'TotalSpent' : np.float}) # преобразуем тип TotalSpent в вещественный
        return X

# преобразование категориальных признаков по алгоритму one-hot-encoding и стандартизирование признаков
    
class DataTransformer(BaseEstimator, TransformerMixin):

    def __init__(self, num_cols, cat_cols):
        self.num_cols = num_cols
        self.cat_cols = cat_cols
        return None
    def fit(self, X, y=None):
        return self  

    def transform(self, X):
        scaler = StandardScaler()
        data_num_train = scaler.fit_transform(X[self.num_cols]) # стандартизируем числовые признаки для обучения
        data_num_train = pd.DataFrame(data_num_train, columns = self.num_cols, index = X.index) 
        # трансформируем категориальные признаки в числовые методом one-hot-encoding для обучения и теста
        data_cat_train = pd.get_dummies(X[self.cat_cols]) 
        # соединяем полученное
        X = pd.concat([data_cat_train, data_num_train], axis=1, sort = False)
        return X
```

### Генерация pipelime 

```python
pipe = make_pipeline(FeatureTransformer(), DataTransformer(num_cols = num_cols, cat_cols = cat_cols), LogisticRegression())
pipe.get_params()
```

```python
# Создаем сетку обучения для логистической регрессии
grid_search = GridSearchCV(pipe, 
                               {'logisticregression__C': np.array([100, 10, 1, 0.1, 0.01, 0.001]),
                                'logisticregression__penalty' :['l1', 'l2'], 
                                'logisticregression__solver' : ['saga', 'liblinear']
                               },
                               cv=5,
                               scoring='roc_auc'
                               )
```

```python
grid_search.fit(X_train, y_train)
```

#### На тесте результат показал себя неплохо (0.84).
